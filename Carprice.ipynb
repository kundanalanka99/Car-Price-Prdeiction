```python
# now we remove unneccessary columns
# as 'Car_Name' is not required we remove it and make a new set
final=df[['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven', 'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner']]
```

```python
# represents top 5 records
final.head()
```

```python
# adding a new column with the name 'Current_Year'
# it is added to know the years of the car
final['Current_Year']=2020
```

```python
# represents top 5 records after adding a new column
final.head()
```

```python
# now adding a new column to get the no of years of a car with the name 'No_Year'
final['No_Year']=final['Current_Year']-final['Year']
```

```python
# now representing with new columns
final.head()
```

```python
# after this we can delete un neccessary columns present using drop
# here parameter axis specifies whether to remove row or a column
# if axis is 1 it drops columns and if 0 it drops rows
# here parameter inplace happens at a particular if it is true otherwise false
final.drop(['Current_Year'],axis=1,inplace=True)
final.drop(['Year'],axis=1,inplace=True)
```

```python
# represents top 5 records after droping the columns
final.head()
```

```python
# here we are converting to categoricals
# by doing this in 'Fuel_Type' one type decreases
# and it represents 1 for which it belongs if both gets 0 and 0 then it belongs to the 3rd type
# the above applies to both 'Seller_Type' and 'Transmission'
final=pd.get_dummies(final,drop_first=True)
# drop_first = True means the reference is droped after encoding.
```

```python
# the left types are added to the column names
# resulted in increase of columns
final.head()
```

```python
# finding the correlations of the dataset
# it is only applied to the numerical values
# one feature with respective to all the features present
# represents correlation table
correlation=final.corr()
correlation
```

## Performing Feature Importance

```python
# importing the required libraries
from sklearn.ensemble import ExtraTreesRegressor
# ensemble method helps to include algorithms on randomized decision tree
model=ExtraTreesRegressor()
# used to fit the no of randomized decision trees
model.fit(X,Y)
```

```python
print(model.feature_importances_)
# returns the feature importance 
# ndarray and floating values display
```

```python
# to plot graph of feature importance
graph=pd.Series(model.feature_importances_,index=X.columns)
# using the above array values
# index - Values must be hashable and have the same length as data
# nlargest() function is used to get the first n rows ordered by columns in descending order
graph.nlargest(5).plot(kind='barh')
# horizontal bars are plotted
```

```python
# train_test_split
```

```python
# import required libraries
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
# test_size should be between 0.0 to 1.0
# If train_size is also None, it will be set to 0.25
```

```python
# displaying X_train
X_train
```

```python
# getting the dimensions of X_train
X_train.shape
```

```python
# import the required libraries
from sklearn.ensemble import RandomForestRegressor
# ensemble method helps to include algorithms on randomized decision tree
# RandomForestRegressor is the other one
rf_random=RandomForestRegressor()
```

```python
# Hyperparameter
import numpy as np
n_estimators=[int(x) for x in np.linspace(start=100, stop=1200, num=12)]
# start is start of the sequence
# stop is end of the sequence
# num is no of samples to generate and it is optional
print(n_estimators)
```

```python
# randomized search cv
from sklearn.model_selection import RandomizedSearchCV
```

```python
# no of features to consider at every split
max_features=['auto','sqrt']
# max no of levels in tree
max_depth=[int(x) for x in np.linspace(5, 30, num=6)]
# min no of samples required to split a node
min_samples_split=[2,5,10,15,100]
# min no of samples required at each leaf node
min_samples_leaf=[1,2,5,10]
```

```python
# creating random grid
random_grid={'n_estimators': n_estimators,
            'max_features': max_features,
            'max_depth': max_depth,
            'min_samples_split': min_samples_split,
            'min_samples_leaf': min_samples_leaf}
print(random_grid)
```

```python
# using the random grid to search for best hyperparameters
# creating a model
rf=RandomForestRegressor()
rf_random=RandomizedSearchCV(estimator=rf, param_distributions= random_grid, scoring='neg_mean_squared_error', n_iter=10, cv= 5, verbose= 2, random_state=42, n_jobs=1)
```

```python
rf_random.fit(X_train,Y_train)
```

```python
predictions=rf_random.predict(X_test)
predictions
```

```python
sns.distplot(Y_test-predictions)
```
